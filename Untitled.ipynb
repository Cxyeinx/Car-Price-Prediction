```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
```

```python
df = pd.read_csv("train.csv")
```

```python
df
```

```python
df = df.drop(["ID", "Levy", "Model", "Drive wheels", "Doors", "Wheel"], axis=1)
```

```python
df
```

```python
one_hot_encoding = ["Manufacturer", "Category", "Leather interior", "Fuel type", "Gear box type", "Color"]
scaling = ["Prod. year", "Engine volume", "Mileage", "Airbags"]
```

```python
def process_engine_volume(x):
    x = float("".join([i for i in x if i.isdigit() or i == "."]))
    return x

df["Engine volume"] = df["Engine volume"].apply(process_engine_volume)
```

```python
def process_mileage(x):
    x = float("".join([i for i in x if i.isdigit() or i == "."]))
    x = 10000 * round(x/10000)
    return x

df["Mileage"] = df["Mileage"].apply(process_mileage)
```

```python
ohe_scaler = LabelEncoder()
minmax_scaler = MinMaxScaler()
```

```python
for col in one_hot_encoding:
    df[col] = ohe_scaler.fit_transform(df[col])
    df[col] = minmax_scaler.fit_transform(np.array(df[col]).reshape(-1, 1))
```

```python
df
```

```python
for col in scaling:
    df[col] = minmax_scaler.fit_transform(np.array(df[col]).reshape(-1, 1))
    
```

```python
df
```

```python
def process_price(x):
#     x = float("".join([i for i in x if i.isdigit() or i == "."]))
    x = 1000 * round(x/1000)

    return x

df["Price"] = df["Price"].apply(process_price)
```

```python
df
```

```python
x = np.array(df.drop("Price", axis=1))[:-5].reshape(19232//32, 32, 11)
y = np.array(df["Price"]).reshape(-1, 1)[:-5].reshape(19232//32, 32, 1)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
x_train, x_test, y_train, y_test = torch.Tensor(x_train), torch.Tensor(x_test), torch.Tensor(y_train), torch.Tensor(y_test)
```

```python
print(x_train.shape, y_train.shape)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

class DNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.block = nn.Sequential(
            nn.Linear(11, 128),
            nn.ReLU(),
            nn.Linear(128, 256),
            nn.ReLU(),
            nn.Linear(256, 128),
            nn.ReLU(),
            nn.Linear(128, 32),
            nn.ReLU(),
            nn.Linear(32, 1)
#             nn.Sigmoid()
                                   )
    def forward(self, x):
        return self.block(x)


loss_function = nn.L1Loss()
net = DNN()
optimizer = optim.Adam(net.parameters(), lr=0.001)


for epoch in range(10_000):
    for xx, yy in zip(x_train, y_train):
        outputs = net(xx)
        loss = loss_function(outputs, yy)
        net.zero_grad()
        loss.backward()
        optimizer.step()
    
    if epoch % 1000:
        with torch.no_grad():
            output = net(x_test)
            l = loss_function(output, y_test)
        print(f"training loss: {loss}, testing loss: {l}")
        
```
